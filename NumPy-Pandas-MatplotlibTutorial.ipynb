## Working with large data

Suppose you have a massive dataset that you cannot load alltogether. With pandas you can load it in 'pieces'/'Chunks'

```
for poke in pd.read_csv('modified.csv',chunksize = 10):
  print('CHUNK OF POKE DATASET')
  print(poke)
# chunksize = n : means that n rows are loaded at a time
```

```
new_poke = pd.DataFrame(columns=poke.columns)

for poke in pd.read_csv('modified.csv',chunksize = 10):
  results = poke.groupby(['Type 1']).count()

  new_poke = pd.concat([new_poke,results])
print(new_poke.head(10))  
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```
x=[1,2,3,4,5]
y=[2,4,8,15,13]

#Resize your graph
plt.figure(figsize=(6,4),dpi=80)

#plt.plot(x, y, label='just  a function',color='purple',linestyle='--',
#  linewidth=2,marker='o',markersize=10,markerfacecolor='yellow')

#Shorthand notation : '[color][marker][line]
plt.plot(x,y,'yo--',label='just a function')

#line 2
#select interval we want to plot points at
x2 = np.arange(2,5.5,0.5)

#plot part of the graph as line
plt.plot(x2[:4],x2[:4]**2,'r-',label='x^2')

#plot remainder of graph as dashed line
plt.plot(x2[3:],x2[3:]**2,'r--',label='projection')

plt.title('My first graph',
          fontdict={'fontsize': 15})
#x,y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
#x,y tickmarks np.arange(start,stop,step)
plt.xticks(ticks=np.arange(1,6,1))
plt.yticks(ticks=np.arange(2,30,5))
#add legends (from the labels)
plt.legend(loc='lower right')

#save our graph (dpi is basically pixels per inch. good for savinf high resolution images)
plt.savefig('mygraph.png',dpi=300)

plt.show()
```

### Bar charts

```
labels= ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['o','O','/']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('o')
# bars[1].set_hatch('O')
# bars[2].set_hatch('/')

plt.figure(figsize=(6,4))
plt.show()
```

```
bins =list(range(40,101,10))

plt.hist(fifa['Overall'],
         bins=bins,
         color='#bbf533')
plt.xticks(bins)

plt.title('Fifa players distribution')
plt.ylabel('# of players')
plt.xlabel('skill level')
plt.show()
```

### Pie Charts

```
fifa['Preferred Foot']
```

```
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels= ['Left','Right']
colors = ['#f54633','#f56d33']
#pie needs a list of numbers
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')

plt.title('Foot preference FIFA players')

plt.show()
```

Let's take a look at the weights of the players

```
fifa.Weight.head(5)
```

We want to take the 'lbs' out and only have the numbers.

```
fifa.Weight=[int(x.strip('lbs')) for x in fifa.Weight]

light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0] 
medium = fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0] 
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0] 
```

```
#plt.style.use('ggplot')

weights = [light, light_medium, medium, medium_heavy, heavy]

labels = [ '<125', '125-150', '150-175', '175-200', '>200']
colors = [ '#fc0394','#d703fc','#2803fc','#56fc03','#03fc90']

plt.pie(weights, labels=labels,
        autopct='%.2f%%',
        pctdistance=0.8,
        explode=(.4,.1,0,0,.4))
      
plt.show()
```
